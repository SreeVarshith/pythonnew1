```python
#pie chart of subdivided subjects
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,7,5,3])
y=np.array([5,10,8,1])
plt.plot(x,y,ls=':',color='g')
plt.show()
y = np.array([35, 25, 25, 10])
m=["maths",'physics','chem','social']

plt.pie(y,labels=m)
plt.show() 
```

```python

x = np.array([0,565,546,456,465])
y = np.array([10, 20, 30, 40,56])

plt.subplot(1,2,1)
plt.plot(x,y,color='g',ls='dashdot')
plt.title("marks")
plt.suptitle("X class")

plt.show()
```
